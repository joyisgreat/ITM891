### Problem 1 - Basic Features of Graphs
Here we will create a complete graph of 5 vertices and determine basic features about the graph, such as the number of edges it has, the degree, etc. Refer to the documentation here for help: https://networkx.org/documentation/stable/tutorial.html.



```python
import networkx as nx
import numpy as np
# Enter your answer under each step.
# Create a complete graph of 5 vertices
G = nx.complete_graph(5)

# Plot the graph
nx.draw(G,with_labels=True)

# How many edges in this graph?
print(f"{G.number_of_edges()} edges")

# What is the adjacency matrix for this graph?
print(f'adjacency matrix is \n {nx.adjacency_matrix(G).todense()}')

# What is the degree for every vertex?
print(f'the degree for every vertex is {dict(G.degree())}')

# What are the neighbors of the vertex 0 in this graph?
print(f'the neighbors of the vertex 0 in this graph is {list(G.neighbors(0))}')
```

```python
# Calculate the shortest path between any two vertices (all the pairs in the graph).
list(nx.shortest_path_length(G))
```

```python
# Calculate the distance between each pair of vertices (0 and 1, 1 and 2, 2 and 3, etc.).
list(nx.shortest_path_length(G))
```

```python
# Calculate the diameter of the graph. 
max_distance = 1
for source_node, distances in nx.shortest_path_length(G):
  for target_node, distance in distances.items():
    if distance > max_distance:
      max_distance = distance
print(f"max distance: {max_distance}")
```

```python
# Is the graph connected? Check to see if it is. 
for source_node, distances in nx.shortest_path_length(G):
  for target_node, distance in distances.items():
    if source_node == target_node:
      continue
    else:
      if distance == 0:
        print('The graph is disconnected')
      else:
        print('The graph is connected')
```

```python
# How many connected components are in the graph?
print("1 component")
```

###Problem 2 - Creating graphs with different amounts of vertices

```python
# Create a 3-regular graph including 8 vertices
G = nx.Graph()
for i in range(8):
  print(i)
  G.add_node(i)
  if i < 7:
    G.add_edge(i, i+1)
G.add_edges_from([(0,7),(0,6),(1,5),(2,3),(2,4),(3,7)])
# Draw it
nx.draw(G, with_labels=True, node_size=700, node_color="skyblue")
print(list(G.degree()))
```

```python
import networkx as nx

# Create an empty graph
G = nx.Graph()

# Add 8 nodes to the graph
G.add_nodes_from(range(8))

# Add edges to make the graph 3-regular
G.add_edges_from([(0,1),(0,2),(0,3),(1,2),(1,4),(2,5),(3,4),(3,5),(4,6),(5,7),(6,7)])

nx.draw(G, with_labels=True, node_size=700, node_color="skyblue")
print(list(G.degree()))
```

```python
# Create a 4-regular graph including 5 vertices
G = nx.Graph()
for i in range(5):
  G.add_node(i)
  if i == 0:
    continue
  elif i == 1:
    G.add_edge(i, i-1)
  elif i == 2:
    G.add_edge(i, i-1)
    G.add_edge(i, i-2)
  else:
    G.add_edge(i, i-1)
    G.add_edge(i, i-2)
    G.add_edge(i, i-3)
G.add_edge(0, 4)
# Draw it
nx.draw(G, with_labels=True, node_size=700, node_color="skyblue")
print(list(G.degree()))
```

```python
# Create a 4-regular graph including 6 vertices
G = nx.Graph()
for i in range(6):
  print(i)
  G.add_node(i)
  if i < 4:
    G.add_edge(i, i+1)
    G.add_edge(i, i+2)
G.add_edges_from([(0,5),(0,4),(1,5),(4,5)])

# Draw it
nx.draw(G, with_labels=True, node_size=700, node_color="skyblue")
print(list(G.degree()))
```
